```python
import pandas as pd
import matplotlib.pyplot as plot

df = pd.read_csv("https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv")
```

```python
#basic descriptions of dataframe
df.head()
```

```python
df.columns
```

```python
#replacing all non-numeric values in "Mean Score" column with NaN
df['Mean Score'] = pd.to_numeric(df['Mean Score'], errors="coerce")
```

```python
#removing all non-numeric values (NaN) in the "Mean Score" column
df.dropna(subset=["Mean Score"], inplace=True)
```

```python
#making sure above worked
df.head()
df['Mean Score'].median()
```

```python
#viewing mean score of one school (PS99) in the Common Core Algebra regents in 2015

ps99algebra = df.loc[(df["School Name"] == "P.S. 099 Isaac Asimov") & (df["Regents Exam"] == "Common Core Algebra") & (df["Year"] == 2015)]
ps99algebra.head()
```

```python
#getting the mean score of ps99 algebra in 2015

ps99mean = ps99algebra["Mean Score"].mean()
print(ps99mean)
```

```python
#looking at all schools algebra scores in 2015

algebra2015 = df.loc[(df["Regents Exam"] == "Common Core Algebra") & (df["Year"] == 2015)]
algebra2015.head()
```

```python
#looking at mean score of all schools for algebra regents in 2015
algebra2015mean = algebra2015["Mean Score"].mean()
print(algebra2015mean)
```

```python
#We can see that PS99's mean algebra score was 9 points higher than the rest of the city's in 2015
```

```python
ps99algebra = ps99algebra.sort_values(by="Total Tested", ascending=False)
ps99algebra.plot(x="Total Tested", y="Mean Score")
```

```python
data = [["PS99", 72.5],
        ["Rest of City", 63.2]]

meanscores = pd.DataFrame(data, columns=["School", "Score"])

meanscores.plot(x="School", y="Score",
                kind='bar', width=0.2,
                ylabel="Mean Algebra Score in 2015",
                color='red')
```
